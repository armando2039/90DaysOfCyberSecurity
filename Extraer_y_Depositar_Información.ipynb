CELDA 4
Función para extraer datos desde REST

```python
def obtener_tabla(url):
    """
    Descarga una tabla publicada mediante Oracle REST Data Services
    y la devuelve como un DataFrame de pandas.
    """
    respuesta = requests.get(url)
    respuesta.raise_for_status()

    datos_json = respuesta.json()

    return pd.DataFrame(datos_json["items"])
```

CELDA 14
Crear las tablas destino

```python
def obtener_tipo_oracle(tipo):
    """
    Convierte los tipos de datos de pandas
    a tipos compatibles con Oracle.
    """

    if "int" in str(tipo):
        return "NUMBER"

    elif "float" in str(tipo):
        return "NUMBER"

    else:
        return "VARCHAR2(4000)"


tablas = {
    "DW_CUSTOMERS": df_customers,
    "DW_ORDERS": df_orders,
    "DW_ORDER_ITEMS": df_order_items,
    "DW_PRODUCT_INFO": df_product_info,
    "DW_STATES": df_states
}

for nombre, df in tablas.items():

    try:
        cursor.execute(f"DROP TABLE {nombre} PURGE")
        print(f"{nombre} eliminada.")
    except:
        pass

    columnas = []

    for columna, tipo in zip(df.columns, df.dtypes):
        columnas.append(
            f"{columna} {obtener_tipo_oracle(tipo)}"
        )

    sql = f"""
    CREATE TABLE {nombre}
    (
        {', '.join(columnas)}
    )
    """

    cursor.execute(sql)

    print(f"✅ {nombre} creada.")

connection.commit()
```

CELDA 15
Insertar los datos

```python
for nombre, df in tablas.items():

    columnas = ",".join(df.columns)

    placeholders = ",".join(
        [":" + str(i+1) for i in range(len(df.columns))]
    )

    sql = f"""
    INSERT INTO {nombre}
    ({columnas})
    VALUES
    ({placeholders})
    """

    datos = []

    for fila in df.itertuples(index=False):

        datos.append(tuple(
            None if pd.isna(valor) else valor
            for valor in fila
        ))

    cursor.executemany(sql, datos)

    print(f"✅ {len(datos)} registros insertados en {nombre}")

connection.commit()
```
